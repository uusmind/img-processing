```python
import numpy as np
from numpy.fft import fft

%pylab inline 
from matplotlib import pyplot as plt
```

```python
dt = 0.001;
fs = 1/dt;
f1 = 10;
f2 = 100;
```

```python
t = np.arange(0, 0.25, dt)
```

```python
s1 = np.sin(2*np.pi*f1*t)
s2 = np.sin(2*np.pi*f2*t)
s3 = np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)
```

```python
fig, ax = plt.subplots(1,3, figsize=(12,4))
ax[0].plot(t,s1)
ax[0].set_title('signal 1')

ax[1].plot(t,s2)
ax[1].set_title('signal 2')

ax[2].plot(t,s3)
ax[2].set_title('signal 1 + signal 2')
```

```python
s3_fft = np.fft.fft(s3)

A = np.abs(s3_fft)
N = float(len(t))
f = np.arange(0,N)/N*fs
```

```python
plt.plot(f, A)
```

```python
plt.plot(f[:int(N/2)], A[:int(N/2)])
```
